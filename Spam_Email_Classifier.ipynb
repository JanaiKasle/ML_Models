**Creating NB Model**

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

```python
print(train_x.isnull().sum())
print(test_x.isnull().sum())
```

**Naive Bayes**

```python
from sklearn.impute import SimpleImputer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score


imputer_x = SimpleImputer(strategy='mean')
train_x_imputed = imputer_x.fit_transform(train_x)
test_x_imputed = imputer_x.transform(test_x)

imputer_y = SimpleImputer(strategy='most_frequent')
train_y_imputed = imputer_y.fit_transform(train_y.reshape(-1, 1))
test_y_imputed = imputer_y.transform(test_y.reshape(-1, 1))

train_y_imputed = train_y_imputed.flatten()
test_y_imputed = test_y_imputed.flatten()

mnb = MultinomialNB(alpha=1.9)
mnb.fit(train_x_imputed, train_y_imputed)
y_pred1 = mnb.predict(test_x_imputed)
print("Accuracy Score for Naive Bayes: ", accuracy_score(y_pred1, test_y_imputed))
```

**SVC**

```python
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

imputer_x = SimpleImputer(strategy='mean')
train_x_imputed = imputer_x.fit_transform(train_x)
test_x_imputed = imputer_x.transform(test_x)

imputer_y = SimpleImputer(strategy='most_frequent')
train_y_imputed = imputer_y.fit_transform(train_y.reshape(-1, 1))
test_y_imputed = imputer_y.transform(test_y.reshape(-1, 1))

train_y_imputed = train_y_imputed.flatten()
test_y_imputed = test_y_imputed.flatten()

svc = SVC(C=1.0, kernel='rbf', gamma='auto')
svc.fit(train_x_imputed, train_y_imputed)
y_pred2 = svc.predict(test_x_imputed)
print("Accuracy Score for SVC: ", accuracy_score(y_pred2, test_y_imputed))
```

```python
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Assuming you have loaded your data into train_x, train_y, test_x, and test_y

# Step 1: Handle missing values in input features (X)
# Create an imputer object for input features
imputer_x = SimpleImputer(strategy='mean')  # You can use other strategies as well

# Fit and transform on training data
train_x_imputed = imputer_x.fit_transform(train_x)

# Transform on test data
test_x_imputed = imputer_x.transform(test_x)

# Step 2: Handle missing values in target variable (y)
# Create an imputer object for the target variable
imputer_y = SimpleImputer(strategy='most_frequent')  # You can use other strategies as well

# Fit and transform on training target variable
train_y_imputed = imputer_y.fit_transform(train_y.reshape(-1, 1))

# Transform on test target variable
test_y_imputed = imputer_y.transform(test_y.reshape(-1, 1))

# Flatten the target variables
train_y_imputed = train_y_imputed.flatten()
test_y_imputed = test_y_imputed.flatten()

# Step 3: Train and evaluate RandomForestClassifier model
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(train_x_imputed, train_y_imputed)
y_pred_rf = rf_classifier.predict(test_x_imputed)
print("Accuracy Score for RandomForestClassifier: ", accuracy_score(y_pred_rf, test_y_imputed))
```
